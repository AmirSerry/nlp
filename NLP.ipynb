```python
import boto3
import pandas as pd

client = boto3.client(
    'comprehend',region_name='us-east-1'
)

data = pd.read_csv('./1429_1.csv', nrows=10000)
data.head(10)
```

```python
data.count()
```

```python
for index, row in data.iterrows():
    response = client.detect_dominant_language(
    Text=row['reviews.title']
    )
    lang_code = response['Languages'][0]['LanguageCode']
    response = client.detect_sentiment(
    Text=row['reviews.title'],
    LanguageCode=lang_code
    )
    print('Text: {}'.format(row['reviews.title']))
    print('Sentiment:\t{}'.format(response['Sentiment']))
    print('\n')


```
